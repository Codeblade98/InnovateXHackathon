```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = ('https://newsapi.org/v2/everything?'
       'q=OpenAI&Microsoft&Finance&'
       'language=en&'
       'from=2023-11-25&'
       'sortBy=popularity&'
       'apiKey=7faa073ad63a4263a57f02e00944c296')

response = requests.get(url)
```

```python
response.json()['articles']
```

```python
res = response.json()['articles']
r = res[-1]

url_ = r['url']
res2 = requests.get(url_)

html_doc = res2.text
soup = BeautifulSoup(html_doc, 'html.parser')

fulltext = soup.find_all(class_="premium-content")

for t in fulltext:
  x = t.text
  y = x.split('\nAdvertisement\n\n')
y
''.join(y)
```

```python
s = []
res = response.json()['articles']
for r in res:
  source = r['source']['id']
  if(source != 'engadget' and source!='business-insider'):
    continue

  url_ = r['url']
  res2 = requests.get(url_)

  html_doc = res2.text
  soup = BeautifulSoup(html_doc, 'html.parser')
  if source=='engadget':
    fulltext = soup.find_all(class_="caas-body")[0].text
  elif source == 'business-insider':
    article = soup.find_all(class_="premium-content")
    for t in article:
      x = t.text
      y = x.split('\nAdvertisement\n\n')
    fulltext = ''.join(y)
  s.append(fulltext)
```

```python
data_list = s
columns = ['Context']
```

```python
df = pd.DataFrame(data_list, columns=columns)
df.head()
```
